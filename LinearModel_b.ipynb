```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Making dataset

```python
np.random.seed(2)
x = np.random.rand(1000)

y = 5*x + 15
y = y + np.random.normal(0, 0.5, y.shape)

plt.scatter(x, y)
```

# Model

```python
def loss(y, y_pred):
  return np.sum(y-y_pred) / len(y)

def model(x, y, alpha):
  np.random.seed(2)
  w = np.random.rand()
  b = np.random.rand()
  losses = []
  dw = 0
  db = 0
  
  for i in range(10000):
    y_pred = w*x + b
    losses.append(loss(y, y_pred))
    
    dw = (-1 / len(y)) * np.sum(x)
    db = -1 
    w -= alpha*dw
    b -= alpha*db

  return w, b, losses
```

# Analysis

```python
theta_1, theta_0, losses = model(x, y, 0.0014)

figure, axis = plt.subplots(1, 2)

axis[0].plot(range(10000), losses)
axis[0].set_title('loss funtion')

y_pred = theta_1*x + theta_0
axis[1].scatter(x, y)
axis[1].plot(x, y_pred, color='red')
axis[1].set_title('curve fitting')

print('Final loss: ', loss(y, y_pred))
```
